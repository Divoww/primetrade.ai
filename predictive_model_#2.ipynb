### Converting Timestamps and aligning datasets by date

```python
df1['date'] = pd.to_datetime(df1['date']).dt.normalize()
df2['date'] = pd.to_datetime(df2['Timestamp IST'], format="%d-%m-%Y %H:%M").dt.normalize()
```

```python
df_merged = df2.merge(df1[['date','classification']], on='date', how='left')
df_merged.head()
```

### Daily PnL and Number of Trades of each Trader

```python
# Daily pnl per trader
daily_pnl = df_merged.groupby(['Account','date','classification'])['Closed PnL'].sum().reset_index(name='total_pnl')

# Number of trades per trader
daily_trades = df_merged.groupby(['Account','date']).size().reset_index(name='num_trades')

daily_trader_data = daily_pnl.merge(daily_trades, on=['date','Account'])
daily_trader_data
```

# Predictive Model #2

### Features and Labels

```python
features = [ 'classification', 'rolling_win_rate_5d', 'rolling_mean_pnl_5d', 'rolling_std_pnl_5d', 'rolling_avg_trades_5d', 'yesterday_pnl' ]
labels = ['target']

# Sorting Values
daily_trader_data = daily_trader_data.sort_values(['Account','date'])

# rolling_win_rate_5d
daily_trader_data['win'] = daily_trader_data['total_pnl'].transform(lambda x: 1 if x>0 else 0)
daily_trader_data['rolling_win_rate_5d'] = daily_trader_data.groupby('Account')['win'].transform(lambda x: x.rolling(5, min_periods=1).mean())
# Note: Here, a no profit no loss day is considered as no win day, unlike elsewhere

# rolling_mean_pnl_5d
daily_trader_data['rolling_mean_pnl_5d'] = daily_trader_data.groupby(['Account'])['total_pnl'].transform(lambda x: x.rolling(5, min_periods=1).mean())

# rolling_std_pnl_5d
daily_trader_data['rolling_std_pnl_5d'] = daily_trader_data.groupby('Account')['total_pnl'].transform(lambda x: x.rolling(5, min_periods=1).std())

# rolling_avg_trades_5d
daily_trader_data['rolling_avg_trades_5d'] = daily_trader_data.groupby('Account')['num_trades'].transform(lambda x: x.rolling(5, min_periods=1).mean())

# yesterday_pnl
daily_trader_data['yesterday_pnl'] = daily_trader_data.groupby('Account')['total_pnl'].shift(1)

# target
daily_trader_data['target'] = daily_trader_data.groupby('Account')['win'].shift(-1)

# dropping missing values
daily_trader_data.dropna(inplace=True)
```

### Data Partition

```python
cutoff_date = daily_trader_data['date'].quantile(0.8)

train = daily_trader_data[daily_trader_data['date']<=cutoff_date]
test = daily_trader_data[daily_trader_data['date']>cutoff_date]

x_train = train[features]
x_test = test[features]
y_train = train[labels]
y_test = test[labels]

print(cutoff_date)
```

### Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler, OrdinalEncoder, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

num_attrbs = [ 'rolling_win_rate_5d', 'rolling_mean_pnl_5d', 'rolling_std_pnl_5d', 'rolling_avg_trades_5d', 'yesterday_pnl' ]
cat_attrbs = ['classification']

X_preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_attrbs),
    ('cat', OrdinalEncoder(), cat_attrbs)
])
x_train = X_preprocessor.fit_transform(x_train)
x_test = X_preprocessor.fit_transform(x_test)

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.fit_transform(y_test)
```

### Training and Prediction

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr = lr.fit(x_train, y_train)
predictions = lr.predict(x_test)

from sklearn.metrics import roc_auc_score
score = roc_auc_score(y_test, predictions)
print(score)
```
