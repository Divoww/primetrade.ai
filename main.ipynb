## Converting Timestamps and aligning datasets by date

```python
df1['date'] = pd.to_datetime(df1['date']).dt.normalize()
df2['date'] = pd.to_datetime(df2['Timestamp IST'], format="%d-%m-%Y %H:%M").dt.normalize()
```

```python
df_merged = df2.merge(df1[['date','classification']], on='date', how='left')
df_merged.head()
```

## Key Metrics

```python
print(len(set(df2["Account"])))
```

### Daily PnL and Number of Trades of each Trader

```python
# Daily pnl per trader
daily_pnl = df_merged.groupby(['Account','date','classification'])['Closed PnL'].sum().reset_index(name='total_pnl')

# Number of trades per trader
daily_trades = df_merged.groupby(['Account','date']).size().reset_index(name='num_trades')

daily_trader_data = daily_pnl.merge(daily_trades, on=['date','Account'])
daily_trader_data
```

### Historical Information of each Account

```python
# Determining Win Rate
closed_trades = df_merged[df_merged['Closed PnL'] != 0]
closed_trades['win'] = closed_trades['Closed PnL'] > 0
win_rate = closed_trades.groupby('Account')['win'].mean().sort_values().reset_index(name='win_rate')

# Other trader aggregates
trader_summary = df_merged.groupby('Account').agg(total_pnl = ('Closed PnL', 'sum'), avg_pnl = ('Closed PnL', 'mean'),pnl_std = ('Closed PnL', 'std'), num_trades = ('Closed PnL', 'count'), avg_trade_size = ('Size Tokens', 'median')).reset_index()

# Merging into single trader_info DataFrame
trader_summary = win_rate.merge(trader_summary, on='Account')

trader_summary.head()
```

### Daily Key Metrics

```python
# Daily long_short_ratio
daily_ls = (
    df_merged[df_merged['Direction'].isin(['Open Long', 'Open Short'])]
    .groupby(['date', 'Direction'])
    .size()
    .unstack(fill_value=0)
)

daily_ls['long_short_ratio'] = (daily_ls['Open Long'] / daily_ls['Open Short'])
daily_ls = daily_ls[daily_ls['long_short_ratio']!=np.inf]

# Daily win_rate
closed_trades = df_merged[df_merged['Closed PnL'] != 0]
closed_trades['win'] = closed_trades['Closed PnL'] > 0
daily_winrate = closed_trades.groupby('date')['win'].mean().reset_index(name='win_rate')

# Daily avg_trade_size
daily_avg_trade_size = df_merged.groupby('date')['Size Tokens'].median().reset_index(name='avg_trade_size')

# Daily num_trades
daily_num_trades = df_merged.groupby('date')['Size Tokens'].size().reset_index(name='num_trades')

# Daily total_pnl
daily_total_pnl = df_merged.groupby('date')['Closed PnL'].sum().reset_index(name='total_pnl')

# Merging these 5 metrics
daily_data = daily_ls.merge(daily_winrate, on='date').merge(daily_avg_trade_size, on='date').merge(daily_total_pnl, on='date').merge(daily_num_trades,on='date')

# Mapping classification to these 4 metrics
daily_data = df1.merge(daily_data,on='date')[['date','classification','long_short_ratio','win_rate','avg_trade_size','total_pnl','num_trades']]

daily_data
```

# Data Analysis

## Does performance differ between Fear vs Greed days?

### Fear vs Greed impact on pnl

```python
daily_data.groupby('classification')['total_pnl'].mean().sort_values()
```

##### -During Extreme Fear, traders may trade more aggressively/frequently  
#####  (which is why the total daily profits are highest during Extreme Fear)

```python
df_merged.groupby('classification')['Closed PnL'].mean().sort_values()
```

##### -During Extreme Greed, trades are more efficient but possibly fewer
#####  (Average profit per trade is highest during Extreme Greed)

```python
# Number of trades by Market Sentiment
daily_data.groupby('classification')['num_trades'].median().sort_values()
```

##### Conclusion
##### Extreme Fear = most trades, highest total daily profits
##### Extreme Greed = lesser trades, highest profit per trade

### Fear vs Greed impact on Win Rate

```python
daily_data.groupby('classification')['win_rate'].mean().sort_values()
```

##### Extreme Fear = most trades, lowest win rate, highest total daily profits (some trades are highly profitable)
##### Extreme Greed = lesser trades, higher win rate, highest profit per trade (consistent profits)

### Fear vs Greed impact on drawdown proxy

```python
# Computing Drawdown Proxy
daily_data = daily_data.sort_values('date')
daily_data['cum_pnl'] = daily_data['total_pnl'].cumsum()
daily_data['rolling_max'] = daily_data['cum_pnl'].cummax()
daily_data['drawdown'] = daily_data['cum_pnl'] - daily_data['rolling_max']
```

```python
daily_data.groupby('classification')['drawdown'].min().sort_values()
```

##### Worst Drawdowns were during Greed.
##### Least worst Drawdowns were during Fear.

```python
daily_data.groupby('classification')['drawdown'].mean().sort_values()
```

##### Average underwater depth was most during Greed and Extreme Fear.
##### Average underwater depth was least during Fear and Extreme Greed.

#### CONCLUSION - Greed has worse drawdowns than Extreme Greed.

## Trader Segmentation

### Frequent vs Infrequent

```python
median_trades = trader_summary['num_trades'].median()

trader_summary['frequency_segment'] = trader_summary['num_trades'].apply(lambda x: 'frequent' if  x > median_trades else 'infrequent')

trader_summary.groupby('frequency_segment')[['total_pnl','win_rate']].mean()
```

##### Infrequent traders have a slightly higher win rate, 
##### But average total PnL is significantly higher for frequent traders.

### Consistent vs Inconsistent

```python
mean_winrate = trader_summary['win_rate'].mean()

trader_summary['consistency_segment'] = trader_summary['win_rate'].apply(lambda x: 'consistent' if  x > mean_winrate else 'inconsistent')

trader_summary.groupby('consistency_segment')[['total_pnl','num_trades','avg_pnl']].mean()
```

##### Inconsistent traders usually trade more, have lower average profits, but higher total profits
##### Consistent traders usually trade less, have higher average profits, but lower total profits

### High Risk vs Low Risk (Leverage Proxy)

```python
median_size = trader_summary['avg_trade_size'].median()

trader_summary['risk_segment'] = trader_summary['avg_trade_size'].apply(lambda x: 'High Risk' if x > median_size else 'Low Risk')

trader_summary.groupby('risk_segment')[['total_pnl','win_rate']].mean()
```

##### Risk averse traders have better total profits than risk takers who have slighly better win rates
