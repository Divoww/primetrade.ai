### Converting Timestamps and aligning datasets by date

```python
df1['date'] = pd.to_datetime(df1['date']).dt.normalize()
df2['date'] = pd.to_datetime(df2['Timestamp IST'], format="%d-%m-%Y %H:%M").dt.normalize()
```

```python
df_merged = df2.merge(df1[['date','classification']], on='date', how='left')
df_merged.head()
```

### Daily PnL and Number of Trades of each Trader

```python
# Daily pnl per trader
daily_pnl = df_merged.groupby(['Account','date','classification'])['Closed PnL'].sum().reset_index(name='total_pnl')

# Number of trades per trader
daily_trades = df_merged.groupby(['Account','date']).size().reset_index(name='num_trades')

daily_trader_data = daily_pnl.merge(daily_trades, on=['date','Account'])
daily_trader_data
```

### Historical Information of each Account

```python
# Determining Win Rate
closed_trades = df_merged[df_merged['Closed PnL'] != 0]
closed_trades['win'] = closed_trades['Closed PnL'] > 0
win_rate = closed_trades.groupby('Account')['win'].mean().sort_values().reset_index(name='win_rate')

# Other trader aggregates
trader_summary = df_merged.groupby('Account').agg(total_pnl = ('Closed PnL', 'sum'), avg_pnl = ('Closed PnL', 'mean'),pnl_std = ('Closed PnL', 'std'), num_trades = ('Closed PnL', 'count'), avg_trade_size = ('Size Tokens', 'median')).reset_index()

# Merging into single trader_info DataFrame
trader_summary = win_rate.merge(trader_summary, on='Account')

trader_summary.head()
```

### Trader Segmentation

#### Frequent vs Infrequent

```python
median_trades = trader_summary['num_trades'].median()

trader_summary['frequency_segment'] = trader_summary['num_trades'].apply(lambda x: 'frequent' if  x > median_trades else 'infrequent')
```

#### Consistent vs Inconsistent

```python
mean_winrate = trader_summary['win_rate'].mean()

trader_summary['consistency_segment'] = trader_summary['win_rate'].apply(lambda x: 'consistent' if  x > mean_winrate else 'inconsistent')
```

#### High Risk vs Low Risk (Leverage Proxy)

```python
median_size = trader_summary['avg_trade_size'].median()

trader_summary['risk_segment'] = trader_summary['avg_trade_size'].apply(lambda x: 'High Risk' if x > median_size else 'Low Risk')
```

# Predictive Model #1

##### This model is based on trader's long run characteristics. 
##### To predict whether a trader will make a profit on a particular day or not, we only need to provide   
##### (1) trader characteristics (frequent/infrequent, consistent/inconsistent, High Risk/Low Risk) and  
##### (2) market sentiment on that day (Extreme Fear/Fear/Neutral/Greed/Extreme Greed)

### Features and Labels

```python
features = ['classification','frequency_segment','consistency_segment','risk_segment']
labels = ['profit?']

columns_unprepared = ['classification','frequency_segment','consistency_segment','risk_segment','total_pnl_x']

main_data = daily_trader_data.merge(trader_summary, on='Account', how='left')[columns_unprepared]
main_data['profit?'] = main_data['total_pnl_x'].apply(lambda x: 'y' if x>0 else 'n')

X = main_data[features]
y = main_data[labels]
```

### Handling Categorical Variables

```python
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder

oe = OrdinalEncoder()
X_encoded = oe.fit_transform(X)

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

### Model Selection

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
# Testing Logistic Regression Model
lr = LogisticRegression()
lr_scores = cross_val_score(estimator=lr, X=X_encoded, y=y_encoded, cv=10, scoring='roc_auc')
print(f"The mean of Logistic Regression ROC-AUC scores is {lr_scores.mean()}")
```

```python
# Testing Decision Tree Classifier Model
cdt = DecisionTreeClassifier(random_state=42)
cdt_scores = cross_val_score(estimator=cdt, X=X_encoded, y=y_encoded, cv=10, scoring='roc_auc')
print(f"The mean of Decision Tree Classifier ROC-AUC scores is {cdt_scores.mean()}")
```

```python
# Testing Random Forest Classifier Model
crf = RandomForestClassifier(random_state=42)
crf_scores = cross_val_score(estimator=crf, X=X_encoded, y=y_encoded, cv=10, scoring='roc_auc')
print(f"The mean of Random Forest Classifier ROC-AUC scores is {crf_scores.mean()}")
```

#### Logistic Regression has highest mean ROC-AUC score (~0.62)

### Data Partition

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)
```

### Training and Prediction

```python
lr = lr.fit(x_train, y_train)
predictions = lr.predict(x_test)

from sklearn.metrics import roc_auc_score
score = roc_auc_score(y_test, predictions)
print(score)
```

##### The ROC-AUC Score of our model is ~0.59
